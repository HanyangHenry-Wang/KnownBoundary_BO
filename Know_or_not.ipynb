```python
import  os
import numpy as np
import matplotlib.pyplot as plt
```

```python
BO_EI = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_BO_EI'), delimiter=',')
logBO_TEI2 = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_logBO_EI'), delimiter=',')
logBO_no_boundary_HanyangEI = np.genfromtxt(os.path.join(os.getcwd(),'exp3/Branin2D_logBO_no_boundary_EI'), delimiter=',')

fstar = 0.397887
N = 20

total_record = np.log10(np.array(BO_EI)-fstar)
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='GP-EI',color='blue')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='blue')

total_record = np.log10(np.array(logBO_no_boundary_HanyangEI)-fstar)
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='logGP-EI',color='red')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='red')

total_record = np.log10(np.array(logBO_TEI2)-fstar)
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='logGP*-EI',color='yellow')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='yellow')




plt.xlabel('iteration')
plt.ylabel('log regret')

plt.legend()
plt.title('Branin2D')
```

# Why log-GP works better? (take Beale 2D and Branin 2D as an example)

A fitst thing that we need to check is whether the advantage is coming from the model training part because log-GP has three hyper-parameters and thus has more restart time. To check it, we set the restart number of the GP training and log-GP totally the same. As we can see, the advantage of log-GP remains. Therefore, this advantage is not because of the number of restart in training but the model itself.

```python
BO_EI = np.genfromtxt(os.path.join(os.getcwd(),'training_check/Beale2D_BO_EI'), delimiter=',')
logBO_no_boundary_HanyangEI = np.genfromtxt(os.path.join(os.getcwd(),'exp3/Beale2D_logBO_no_boundary_EI'), delimiter=',')

fstar = 0.
N = 20

total_record = np.log10(np.array(BO_EI)-fstar)
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='GP-EI',color='blue')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='blue')

total_record = np.log10(np.array(logBO_no_boundary_HanyangEI)-fstar)
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='logGP-EI',color='red')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='red')



plt.xlabel('iteration')
plt.ylabel('log regret')

plt.legend()
plt.title('Beale2D')
```
