```python
import  os
import numpy as np
import matplotlib.pyplot as plt
```

```python
BO_EI = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_BO_EI'), delimiter=',')
BO_MES = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_BO_MES'), delimiter=',')
BO_ERM = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_BO_ERM'), delimiter=',')
BO_TEI = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_BO_TEI'), delimiter=',')
Warped_BO_TEI2 = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_logBO_TEI2'), delimiter=',')

fstar = 0.397887
N = 20

total_record = (np.array(BO_EI)-fstar)[:N,:]
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='GP-EI',color='blue')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='blue')

total_record = (np.array(BO_MES)-fstar)[:N,:]
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='GP-MES*',color='#c22f2f')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='#c22f2f')

total_record = (np.array(BO_TEI)-fstar)[:N,:]
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='GP-TEI',color='green')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='green')

total_record = (np.array(BO_ERM)-fstar)[:N,:]
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='transformed GP-ERM',color='black')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='black')

total_record = (np.array(Warped_BO_TEI2)-fstar)[:N,:]
mean = np.mean(total_record,axis=0)
std = np.std(total_record,axis=0)
iter = np.array(range(mean.shape[0]))
plt.plot(mean,linewidth=3,label='logGP-TEI',color='yellow')
plt.fill_between(iter, mean-std/np.sqrt(N), mean+std/np.sqrt(N),alpha=0.25,color='yellow')


plt.xlabel('iteration')
plt.ylabel('regret')

plt.legend()
plt.title('Branin2D')
```
